## 1 Brownian Motion in 1D

### 1.1 The scenario

```python
# in 1D(line)
#below
# probability of going right as p
#going left is q(=1p)
```

### 1.2 The simulation

```python
def brown_1d(prob_right=.5, no_of_steps=10000):
    ''' 
        Returns the final position of the particle 
        after the number of steps.
    '''

    step_size = 1    
    x = 0    # starting x position

    for _ in range(no_of_steps):
        if rnd.random() < prob_right:   
            x += step_size
        else:
            x -= step_size

    return x
```

```python
import numpy.random as rnd
```

```python
no_of_steps, p = 1_000, .5
repeats = 10_000
results = [brown_1d(prob_right=p, no_of_steps=no_of_steps)
           for _ in range(repeats)]

plt.hist(results, bins=25, density=True)
plt.xlabel(f'Distance from the origin after {no_of_steps} steps')
plt.title(f'Probability distribution for 1D Random walk with p={p}')
plt.xkcd()
plt.show()
```

### 1.3 What does theory say?

```python
def normal(x, m=0, s=1):
    s2 = s**2
    output = np.exp(-(x-m)**2/(2*s2))
    output /= np.sqrt(2*np.pi*s2)
    return output
```

```python
no_of_steps, p = 1_000, .5
repeats = 10_000
results = [brown_1d(prob_right=p, no_of_steps=no_of_steps)
           for _ in range(repeats)]

plt.hist(results, bins=25, density=True)
plt.xlabel(f'Distance from the origin after {no_of_steps} steps')
plt.title(f'Probability distribution for 1D Random walk with p={p}')

prob_right = p
prob_left = 1 - prob_right
mean = no_of_steps * (prob_right - prob_left)
sd = np.sqrt(4*no_of_steps*prob_right*prob_left)
x = np.unique(results)

plt.plot(x, normal(x, mean, sd), label='Theoretical')
plt.show()
```

## 2 Estimating \(\pi\)

### 2.1 The method

```python
#randomly creat points by using NumPys PRNG
#use if the point is inside or outside to get area of the cirle
#--> get an estimate for 
#pi= 4* N/Ntotal
```

### 2.2 A poor but intuitive solution

```python
N_total = 100_000 
N_green = 0

for _ in range(N_total):
    x = rnd.uniform(low=-1, high=1, size=1)
    #
    y = rnd.uniform(low=-1, high=1, size=1)
    r = np.sqrt(x**2 + y**2)

    if r <= 1:
        N_green += 1

4 * N_green / N_total
```

### 2.3 A better solution

```python
N_total=1_000_000
x=rnd.uniform(low=-1, high=1, size=N_total)
y=rnd.uniform(low=-1, high=1, size=N_total)
N_green= np.sum((x**2+y**2) <= 1)

4 * (N_green / N_total)
```

### 2.4 A smarter solution

```python
N_total=1_000_000
x=rnd.rand(N_total)
y=rnd.rand(N_total)
N_green=np.sum((x**2+y**2) <= 1)

4 * (N_green / N_total)
```

```python
#x and y are both [0,1)
#rand is faster
```

## Exercise 1 :  Vital Stats of Random 

```python
import numpy.random as rnd

from matplotlib import pyplot as plt

def estimate_pi(n):
    N_total=n
    x=rnd.rand(N_total)
    y=rnd.rand(N_total)
    N_green=np.sum((x**2+y**2) <= 1)

    return 4 * (N_green / N_total)

estimate_pi(n=10_000)
```

```python
n_values = [10,100,1_000,10_000,100_000,1_000_000]

n_std = []
n_mean = []

for n_value in n_values:
    
    estimate_pi(n= n_value)
    no_of_repeats = 100
    pi_results = [estimate_pi(n_value) for _ in range(no_of_repeats)]
    n_std.append(np.std(pi_results))
    n_mean.append(np.mean(pi_results))
```

```python
x = n_values
y = n_mean

plt.xscale('log')
plt.hlines (np.pi,  x[0], x[-1], linestyle='--', colors= 'grey',
            label = 'Real $pi$')

plt. errorbar(x, y, xerr = 6, yerr = n_std,
              fmt = 'o', capsize = 0, 
              label = 'Std. Dev. of $pi$ values')
              
plt.legend (loc = 'upper right')
              
plt.xlabel('No. of points used to generate value')
plt.ylabel ('Value of $pi$')
plt.title('Value of $pi$ vs. No. of points used') 

plt.show()
```

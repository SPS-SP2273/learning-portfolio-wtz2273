```python
def pool(list_of_id):
    combined_file_content_2 = ""
    for ids in list_of_id:
        with open(f'person_{ids:05}.txt', 'r') as file:
            indi_file_content = file.read()
        combined_file_content_2 = combined_file_content_2 + indi_file_content
    return (combined_file_content_2)
```

```python
def chunk(L, n):
    for i in range(0, len(L), n):
        yield L[i:i+n]
```

```python
l = [s for s in range(100)]
n = 10
for num_list in chunk(l,n):
    id_list= num_list
    print (test(pool(id_list)))
```
